```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nitrogeno_suelo_anual - Hoja 1.csv')
```

```python
df
```

```python
x = df.mes
y = df.nitrogeno
plt.plot(x,y)
plt.xlabel("Mes")
plt.ylabel("Nitrgeno del suelo")
plt.title("contenido de nitrgeno del suelo(Tiempo)")
plt.xticks(rotation=90)
plt.show()
```
